```python
import warnings
warnings.filterwarnings('ignore')
import re
import urllib.request
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from requests import get
import requests
import time
from tqdm import tqdm
```

```python
search_query="mobile"

base_url="https://www.amazon.in/s?k="


url=base_url+search_query
url

header={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}

search_response=requests.get(url,headers=header)
search_response.status_code

search_response.text
```

```python
cookie={} # insert request cookies within{}
def getAmazonSearch(search_query,page_number):
    url="https://www.amazon.in/s?k="+search_query+"&page="+page_number
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"

def Searchasin(asin):
    url="https://www.amazon.in/dp/"+asin
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
    
    
def Searchreviews(review_link):
    url="https://www.amazon.in"+review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
product_names=[]
for page in range(0,3):
    response=getAmazonSearch('mobile',str(page))
    soup=BeautifulSoup(response.text, 'lxml')
    for i in soup.findAll("span",{'class':'a-size-medium a-color-base a-text-normal'}): # the tag which is common for all the names of products
        product_names.append(i.text) #adding the product names to the list
product_names
```

```python
data_asin=[]
for page in range(0,3):
    response=getAmazonSearch('mobile',str(page))
    soup=BeautifulSoup(response.text, 'lxml')
    for i in soup.findAll("div",{'class':"sg-col-20-of-24 s-result-item s-asin sg-col-0-of-12 sg-col-28-of-32 sg-col-16-of-20 sg-col sg-col-32-of-36 sg-col-12-of-16 sg-col-24-of-28"}):
        data_asin.append(i['data-asin'])
```

```python
len(data_asin)
```

```python
link=[]
for i in range(len(tqdm(data_asin)):
    response=Searchasin(data_asin[i])
    soup=BeautifulSoup(response.text, 'lxml')
    for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
        link.append(i['href'])  
```

```python
len(link)
```

```python
review = {} 
review['product_names']  = []  
review['review'] = []   
for j in tqdm(range(len(link))):
    for k in range(100):
        response=Searchreviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.text, 'lxml')
        for i in soup.findAll("span",{'data-hook':"review-body"}):
            review['product_names'].append(link[j])
            review['review'].append(i.text)
        
```

```python
df_final=pd.DataFrame.from_dict(review)

df_final.to_csv('Amazon_Reviews.csv')   
```
